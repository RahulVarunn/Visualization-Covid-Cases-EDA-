```python
case=data.loc[data['Total Cases']==data['Total Cases'].max(),'State/UTs']
active=data.loc[data['Active']==data['Active'].max(),'State/UTs']
dis=data.loc[data['Discharged']==data['Discharged'].max(),'State/UTs']
Deaths=data.loc[data['Deaths']==data['Deaths'].max(),'State/UTs']
population=data.loc[data['Population']==data['Population'].max(),'State/UTs']




print("Total Maximum number of cases : ",case.to_string(index=False),"(",data['Total Cases'].max(),")")
print("Total Maximum number of Active cases : ",active.to_string(index=False),"(",data['Active'].max(),")")
print("Highest Death Rate : ",Deaths.to_string(index=False),"(",data['Deaths'].max(),")")
print("Highest population Rate : ",Deaths.to_string(index=False),"(",data['Population'].max(),")")
```

```python
case=data.loc[data['Total Cases']==data['Total Cases'].min(),'State/UTs']
active=data.loc[data['Active']==data['Active'].min(),'State/UTs']
dis=data.loc[data['Discharged']==data['Discharged'].min(),'State/UTs']
Deaths=data.loc[data['Deaths']==data['Deaths'].min(),'State/UTs']
population=data.loc[data['Population']==data['Population'].min(),'State/UTs']




print("Total Maximum number of cases : ",case.to_string(index=False),"(",data['Total Cases'].min(),")")
print("Total Maximum number of Active cases : ",active.to_string(index=False).split(),"(",data['Active'].min(),")")
print("Highest Death Rate : ",Deaths.to_string(index=False),"(",data['Deaths'].min(),")")
print("Highest population Rate : ",Deaths.to_string(index=False),"(",data['Population'].min(),")")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(40,20),clear=10)
plt.bar(data['State/UTs'],data['Total Cases'],label='State\n1 cases = 10 lakhs cases',color='orange')
plt.xticks(rotation=45,fontsize=20)
plt.yticks(fontsize=20)
plt.legend(fontsize=30)
plt.xlabel('STATE',fontsize=30)
plt.ylabel('\nCases\n',fontsize=30)
plt.title("\n\nTOTAL CASES\n",fontsize=30)

plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(40,20),clear=10)
plt.plot(data['State/UTs'],data['Death Ratio'],label='Death ratio',color='orange')
plt.xticks(rotation=45,fontsize=20)
plt.yticks(fontsize=20)
plt.legend(fontsize=30)
plt.xlabel('STATE',fontsize=30)
plt.ylabel('\nDeath Ratio\n',fontsize=30)
plt.title("\nDeath Ratio\n",fontsize=30)
plt.show()

plt.figure(figsize=(40,20),clear=10)
plt.plot(data['State/UTs'],data['Discharge Ratio'],label='Discharge ratio',color='orange')
plt.xticks(rotation=45,fontsize=20)
plt.yticks(fontsize=20)
plt.legend(fontsize=30)
plt.xlabel('State',fontsize=30)
plt.ylabel('\nDischarge ratio\n',fontsize=30)
plt.title("\nDischarge ratio\n",fontsize=30)
plt.show()


plt.figure(figsize=(40,20),clear=10)
plt.plot(data['State/UTs'],data['Active Ratio'],label='Active Ratio',color='orange')
plt.xticks(rotation=45,fontsize=20)
plt.yticks(fontsize=20)
plt.legend(fontsize=30)
plt.xlabel('STATE',fontsize=30)
plt.ylabel('\nActive Ratio\n',fontsize=30)
plt.title("\nActive CASES\n",fontsize=30)
plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.arange(36)
fig = plt.figure(figsize=(20,10))
ax = fig.add_axes([0,0,1,1])
ax.bar(X + 0.00, data['Total Cases'], color = 'orange', width = 0.25,label='(1case = 10 Lakh cases )Total case ')
ax.bar(X + 0.25, data['Discharged'], color = 'pink', width = 0.25,label='Discharge')
ax.bar(X + 0.50, data['Deaths'], color = 'skyblue', width = 0.25,label='Deaths')
plt.xticks(X,data['State/UTs'].to_list(),rotation=45)
plt.xlabel('States')
plt.ylabel('Cases')
plt.title('COVID CASES')
plt.legend()
plt.show()
```
